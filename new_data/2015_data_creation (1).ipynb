Determining player performance value:
Add win %

```python
team_data = pd.read_csv('team_stats_filtered.csv')
team_data.head()
```

```python
before_merge = len(df_15)

# Perform the merge
df_with_win_percent = pd.merge(team_data, df_15, left_on='team_long_name', right_on='teams_played_for', how='inner')

# Number of rows after merge
after_merge = len(df_with_win_percent)

# Calculate how many were dropped
dropped_rows = before_merge - after_merge

print(f'Rows before merge: {before_merge}')
print(f'Rows after merge: {after_merge}')
print(f'Rows dropped (no matching team): {dropped_rows}')
```

```python
df_with_win_percent.head()
```

```python
df_with_win_percent['points_per_game'] = (
    (df_with_win_percent['wins'] * 3 + df_with_win_percent['ties']) / df_with_win_percent['games_played']
)
```

```python
df_with_win_percent.head()
```

```python
df_with_win_percent['position_x'].head()
```

```python
# Record original number of rows
original_rows = len(df_with_win_percent)

# Drop rows where 'market_value_in_eur' is 0 or NaN
df_with_win_percent = df_with_win_percent[df_with_win_percent['market_value_in_eur'].fillna(0) != 0]

# Record how many rows were dropped
dropped_rows = original_rows - len(df_with_win_percent)
print(f"Rows dropped due to zero or missing market value: {dropped_rows}")

# Define PPV calculation
def calculate_ppv(row):
    assists = row['assists'] if pd.notna(row['assists']) else 0
    goals = row['goals'] if pd.notna(row['goals']) else 0
    win_pct = row['points_per_game'] if pd.notna(row['points_per_game']) else 0
    value = row['market_value_in_eur']
    position = str(row['position_x'])

    # Now no need to check for zero value, already filtered
    denominator = 0.30 * value

    if 'F' in position:
        numerator = 0.40 * goals + 0.20 * assists + 0.10 * win_pct + 0.30 * value
    else:
        numerator = 0.30 * assists + 0.30 * goals + 0.10 * win_pct + 0.30 * value

    return numerator / denominator

# Apply the function
df_with_win_percent['player_performance_valuation'] = df_with_win_percent.apply(calculate_ppv, axis=1)
```

```python
df_with_win_percent_sorted = df_with_win_percent.sort_values(by='player_performance_valuation', ascending=False)
```
